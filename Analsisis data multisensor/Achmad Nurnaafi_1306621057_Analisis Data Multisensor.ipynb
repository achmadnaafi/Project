```python
import pandas as pd
import numpy as np
import scipy.stats as stats
```

# 1. Fitur Gradien Terbesar

```python
def calculate_gradmax(data):
    gradmax = np.max(np.abs(np.diff(data, axis=0)), axis=0)
    return gradmax
```

```python
data_frames = []
labels = []
```

```python
sensor_labels = ["F1", "F2", "F3", "F4", "F5", "F6", "F7", "F8", "F9", "F10"]

for i in range(1, 36):
    file_name_A = f"_A_{i}_crop1030.csv"

    data_A = pd.read_csv(file_name_A)
    gradmax_A = calculate_gradmax(data_A[sensor_labels].values)
    data_frames.append(gradmax_A)
    labels.append("A")

for i in range(1, 36):
    file_name_B = f"_B_{i}_crop1030.csv"

    data_B = pd.read_csv(file_name_B)
    gradmax_B = calculate_gradmax(data_B[sensor_labels].values)
    data_frames.append(gradmax_B)
    labels.append("B")
```

```python
column_names = [f"gradmaxF{i+1}" for i in range(len(sensor_labels))]
combined_gradmax_data = pd.DataFrame(data_frames, columns=column_names)
combined_gradmax_data["Zat"] = labels
```

```python
combined_gradmax_data
```

```python
combined_gradmax_data.to_csv("gradmax.csv", index=False)
```

**Analisis**

```python
df = pd.read_csv('gradmax.csv')
```

```python
gradmax_columns = ['gradmaxF1', 'gradmaxF2', 'gradmaxF3', 'gradmaxF4', 'gradmaxF5',
                   'gradmaxF6', 'gradmaxF7', 'gradmaxF8', 'gradmaxF9', 'gradmaxF10']
```

```python
for sensor in gradmax_columns:
    zat_A = df[df['Zat'] == 'A'][sensor]
    zat_B = df[df['Zat'] == 'B'][sensor]

    t_statistic, p_value = stats.ttest_ind(zat_A, zat_B)
    print(f"Uji T-Test ({sensor}):")
    print(f"Statistic: {t_statistic}")
    print(f"P-Value: {p_value}")

    alpha = 0.05
    if p_value < alpha:
        print(f"Terdapat perbedaan signifikan antara jenis Zat A dan B.")
    else:
        print(f"Tidak terdapat perbedaan signifikan antara jenis Zat A dan B")
    print()
```

Dari data yang didapat, kita memiliki nilai gradmax untuk dua zat yang berbeda, yaitu Zat A dan Zat B. Untuk menganalisis apakah gradmax kedua zat bisa dibedakan, kita perlu melakukan uji statistik.

untuk uji statistik saya menggunakan uji t-test, untuk menguji apakah terdapat perbedaan signifikan antara nilai gradmax Zat A dan Zat B untuk masing-masing sensor.

Dari hasil uji t-test untuk nilai gradmax (tingkat perubahan maksimum) pada masing-masing sensor (F1-F10), kita dapat menyimpulkan bahwa:

**GradmaxF1-GradmaxF10:**
   - Nilai p-value yang sangat kecil menunjukkan bahwa ada perbedaan yang signifikan antara jenis Zat A dan B pada semua sensor.
   - Ini berarti, jika kita melihat tingkat perubahan maksimum pada setiap sensor, kita dapat membedakan dengan pasti antara jenis Zat A dan B.

Artinya, respon sensor terhadap Zat A dan B berbeda secara nyata dalam hal tingkat perubahan maksimum pada setiap sensor. Dengan demikian, fitur gradmax bisa dianggap sebagai petunjuk yang kuat untuk membedakan kedua jenis zat dalam dataset ini.







# 2. Fitur Koefisien Fourier Terbesar

```python
def calculate_fourier_max(data):
    fft_result = np.fft.fft(data, axis=0)
    fft_max = np.max(np.abs(fft_result), axis=0)
    return fft_max
```

```python
data_frames = []
labels = []
```

```python
sensor_labels = ["F1", "F2", "F3", "F4", "F5", "F6", "F7", "F8", "F9", "F10"]

for i in range(1, 36):
    file_name_A = f"_A_{i}_crop1030.csv"
    data_A = pd.read_csv(file_name_A)
    fft_max_A = calculate_fourier_max(data_A[sensor_labels].values)
    data_frames.append(fft_max_A)
    labels.append("A")

for i in range(1, 36):
    file_name_B = f"_B_{i}_crop1030.csv"
    data_B = pd.read_csv(file_name_B)
    fft_max_B = calculate_fourier_max(data_B[sensor_labels].values)
    data_frames.append(fft_max_B)
    labels.append("B")
```

```python
column_names = [f"fouriermF{i+1}" for i in range(len(sensor_labels))]
combined_fourier_max_data = pd.DataFrame(data_frames, columns=column_names)
combined_fourier_max_data["Zat"] = labels
```

```python
combined_fourier_max_data
```

```python
combined_fourier_max_data.to_csv("fouriermax.csv", index=False)
```

**Analisis**

```python
fourier_data = pd.read_csv("fouriermax.csv")
```

```python
for i in range(1, 11):
    t_statistic, p_value = stats.ttest_ind(
        fourier_data[fourier_data["Zat"] == "A"][f"fouriermF{i}"],
        fourier_data[fourier_data["Zat"] == "B"][f"fouriermF{i}"]
    )
    print(f"Uji T-test (FouriermF{i}):")
    print(f"Statistic: {t_statistic}")
    print(f"P-Value: {p_value}")

    alpha = 0.05
    if p_value < alpha:
        print(f"Terdapat perbedaan signifikan antara jenis Zat A dan B.")
    else:
        print(f"Tidak terdapat perbedaan signifikan antara jenis Zat A dan B")
    print()
```

Dari hasil uji t-test untuk masing-masing Fourier Max (FouriermF1-FouriermF10), kita dapat membuat analisis sebagai berikut:

**FouriermF1 hingga FouriermF10:**
   - Nilai p-value untuk Fourier Max F1 hingga F10 menunjukkan bahwa sebagian besar kolom memiliki perbedaan signifikan antara jenis Zat A dan B.
   - Terdapat perbedaan yang nyata dalam karakteristik Fourier Max antara kedua jenis zat.

Namun, perlu dicatat bahwa untuk Fourier Max F9, nilai p-value cukup tinggi (lebih besar dari tingkat signifikansi α = 0.05), sehingga tidak dapat dianggap sebagai bukti yang cukup untuk menyatakan adanya perbedaan signifikan antara jenis Zat A dan B.

Secara keseluruhan, hasil uji t-test menunjukkan bahwa Fourier Max dari kedua jenis zat dapat dibedakan secara signifikan untuk sebagian besar kolom Fourier Max F1 hingga F10. Ini dapat diartikan bahwa karakteristik spektrum frekuensi dari respons sensor terhadap jenis Zat A dan B berbeda secara nyata. Meskipun demikian, perlu perhatian khusus pada interpretasi hasil uji t-test dan nilai p-value untuk setiap fitur Fourier Max.
